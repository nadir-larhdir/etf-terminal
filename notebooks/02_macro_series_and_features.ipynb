# Macro Series And Features

Inspect raw macro series from `macro_data` and derived features from `macro_features`.

```python
import matplotlib.pyplot as plt

from db.connection import get_engine
from stores.macro import MacroStore, MacroFeatureStore
```

```python
DATA_BACKEND = "local"
APP_ENV = "uat"
START_DATE = "2024-01-01"
```

```python
engine = get_engine(data_backend=DATA_BACKEND, app_env=APP_ENV)
macro_store = MacroStore(engine)
macro_feature_store = MacroFeatureStore(engine)
```

```python
raw_series = ["DGS2", "DGS10", "BAMLC0A0CM", "BAMLH0A0HYM2"]
raw_matrix = macro_store.get_series_matrix(raw_series, start_date=START_DATE)
raw_matrix.tail()
```

```python
feature_names = ["UST_2S10S", "BEI_5Y", "IG_OAS_LEVEL", "HY_MINUS_IG_OAS", "IG_OAS_Z20"]
feature_matrix = macro_feature_store.get_feature_matrix(feature_names, start_date=START_DATE)
feature_matrix.tail()
```

```python
raw_matrix[["DGS2", "DGS10"]].plot(figsize=(12, 5), title="Treasury Yields")
plt.show()
```

```python
(raw_matrix[["BAMLC0A0CM", "BAMLH0A0HYM2"]] * 100).plot(figsize=(12, 5), title="Credit OAS (bps)")
plt.show()
```
