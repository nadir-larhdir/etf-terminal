# Prices And Returns

Pull ETF prices from ETF Terminal and build simple return series.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from db.connection import get_engine
from stores.market import PriceStore
```

```python
DATA_BACKEND = "local"
APP_ENV = "uat"
TICKERS = ["LQD", "HYG", "JNK", "IEF"]
START_DATE = "2024-01-01"
```

```python
engine = get_engine(data_backend=DATA_BACKEND, app_env=APP_ENV)
price_store = PriceStore(engine)
histories = price_store.get_multi_ticker_price_history(TICKERS, start_date=START_DATE)

prices = pd.DataFrame({ticker: frame["adj_close"] for ticker, frame in histories.items()}).dropna()
returns = np.log(prices).diff().dropna()

prices.tail()
```

```python
returns.tail()
```

```python
(prices / prices.iloc[0] * 100).plot(figsize=(12, 5), title="Normalized Price Index")
plt.show()
```
